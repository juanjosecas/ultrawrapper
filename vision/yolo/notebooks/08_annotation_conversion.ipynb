# 08 – Annotation Format Conversion

Convert between COCO, YOLO, VOC, LabelMe, and Roboflow formats
through a unified internal representation.

```python
from vision.yolo.annotations.convert import convert_annotations
from vision.yolo.annotations.internal import AnnotationSample, Annotation
```

```python
# Build a sample in the internal format
sample = AnnotationSample(
    image_path='cat.jpg',
    width=640,
    height=480,
    annotations=[
        Annotation(
            task='detect',
            class_id=0,
            class_name='cat',
            bbox=[100.0, 80.0, 300.0, 350.0],
        )
    ],
)
print(sample.to_dict())
```
