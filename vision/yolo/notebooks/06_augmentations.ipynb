# 06 – Augmentations

Build Albumentations pipelines with preset strategies without touching YOLO internals.

```python
import numpy as np
from vision.yolo.augmentations import build_augmentation_pipeline, apply_augmentation
```

```python
# Build a medium augmentation pipeline
pipeline = build_augmentation_pipeline(preset='medium', image_size=640)
print(pipeline)
```

```python
# Apply to a dummy image + bounding boxes
image = np.random.randint(0, 255, (640, 640, 3), dtype=np.uint8)
bboxes = [[0.5, 0.5, 0.3, 0.4]]  # YOLO format: [x_c, y_c, w, h] normalised
class_labels = [0]

result = apply_augmentation(pipeline, image, bboxes=bboxes, class_labels=class_labels)
print('Output image shape:', result['image'].shape)
print('Transformed bboxes:', result['bboxes'])
```

```python
# Microscopy preset (heavy augmentation, no colour shift)
import albumentations as A
micro_pipeline = build_augmentation_pipeline(
    preset='microscopy',
    image_size=512,
    custom_transforms=[A.Equalize(p=0.3)],  # add custom transform
)
print('Microscopy pipeline created')
```
