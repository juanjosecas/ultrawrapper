# 06 - Augmentations

Pipelines de Albumentations para imagenes, cajas y keypoints.


```python
from pathlib import Path
import sys

cwd = Path.cwd()
if (cwd / "vision").exists():
    PROJECT_ROOT = cwd
elif len(cwd.parents) >= 3 and (cwd.parents[2] / "vision").exists():
    PROJECT_ROOT = cwd.parents[2]
else:
    PROJECT_ROOT = cwd

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

NOTEBOOK_DIR = PROJECT_ROOT / "vision" / "yolo" / "notebooks"
OUTPUT_DIR = NOTEBOOK_DIR / "outputs"
OUTPUT_DIR.mkdir(exist_ok=True)

IMAGE = str(NOTEBOOK_DIR / "bus.jpg") if (NOTEBOOK_DIR / "bus.jpg").exists() else "https://ultralytics.com/images/bus.jpg"

import cv2
import numpy as np
import matplotlib.pyplot as plt
from vision.yolo.augmentations import build_augmentation_pipeline, apply_augmentation
```

```python
pipeline = build_augmentation_pipeline(preset="medium", image_size=640)
print(pipeline)
```

```python
# Aplicar a imagen dummy + bounding boxes en formato YOLO
image = np.random.randint(0, 255, (640, 640, 3), dtype=np.uint8)
bboxes = [[0.5, 0.5, 0.3, 0.4]]
class_labels = [0]
aug = apply_augmentation(pipeline, image, bboxes=bboxes, class_labels=class_labels)
print(aug.keys())
print(aug["image"].shape, aug.get("bboxes"))
```

```python
# Comparacion visual rapida
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].imshow(image)
axes[0].set_title("Original")
axes[0].axis("off")
axes[1].imshow(aug["image"])
axes[1].set_title("Augmented")
axes[1].axis("off")
fig.tight_layout()
fig.savefig(OUTPUT_DIR / "06_augmentation_preview.png", dpi=150, bbox_inches="tight")
plt.show()
```
